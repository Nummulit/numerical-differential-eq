#### 1. (*Implementation of RK methods*)
Implement RH2, Kutta, and RK4 methods. Numerically verify their order.

```python
def rk2(func, domain, y0):
    ys = [y0]
    for h, x in zip(np.diff(domain), domain):
        k1 = h * func(x, ys[-1])
        k2 = h * func(x + h / 2, ys[-1] + k1 / 2)
        ys.append(ys[-1] + k2)
    return ys
```

```python
def rk4(func, domain, y0):
    ys = [y0]
    for h, x in zip(np.diff(domain), domain):
        k1 = h * func(x, ys[-1])
        k2 = h * func(x + h / 2, ys[-1] + k1 / 2)
        k3 = h * func(x + h / 2, ys[-1] + k2 / 2)
        k4 = h * func(x + h, ys[-1] + k3)
        ys.append(ys[-1] + (k1 + 2 * k2 + 2 * k3 + k4) / 6)
    return ys
```

```python
f = lambda x, y : np.cos(x)
```

```python
x = np.linspace(0, 2 * np.pi, num=3)
```

```python
plt.plot(x, np.sin(x), 'g', label='Exact solution')
plt.plot(x, rk2(f, x, 0), 'k:', label='RK2')
plt.plot(x, rk4(f, x, 0), 'r:', label='RK4')
plt.grid(alpha=0.3)
plt.legend()
plt.show()
```
